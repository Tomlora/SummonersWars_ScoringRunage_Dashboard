```python
import pandas as pd
```

```python
df = pd.read_csv('data_test.csv')

property = {0: 'Aucun',
                1: 'HP',
                2: 'HP%',
                3: 'ATQ',
                4: 'ATQ%',
                5: 'DEF',
                6: 'DEF%',
                8: "SPD",
                9: 'CRIT',
                10: 'DCC',
                11: 'RES',
                12: 'ACC'}

for c in ['innate_type', 'first_sub', 'second_sub', 'third_sub', 'fourth_sub', 'main_type']:
        df[c] = df[c].map(property)
```

```python
df_first = pd.pivot_table(df, index=['first_sub', 'rune_set'], values='first_sub_value', aggfunc='max').reset_index()
df_second = pd.pivot_table(df, index=['second_sub', 'rune_set'], values='second_sub_value', aggfunc='max').reset_index()
df_third = pd.pivot_table(df, index=['third_sub', 'rune_set'], values='third_sub_value', aggfunc='max').reset_index()
df_fourth = pd.pivot_table(df, index=['fourth_sub', 'rune_set'], values='fourth_sub_value', aggfunc='max').reset_index()

df_max = df_first.merge(df_second, left_on=['first_sub', 'rune_set'], right_on=['second_sub', 'rune_set'])
df_max = df_max.merge(df_third, left_on=['first_sub', 'rune_set'], right_on=['third_sub', 'rune_set'])
df_max = df_max.merge(df_fourth, left_on=['first_sub', 'rune_set'], right_on=['fourth_sub', 'rune_set'])


df_max.drop(['second_sub', 'third_sub', 'fourth_sub'], axis=1, inplace=True)
```

```python
df_max
```

```python
df_max['max_value'] = df_max[['first_sub_value', 'second_sub_value', 'third_sub_value', 'fourth_sub_value']].max(axis=1)
df_max.rename(columns={'first_sub' : 'substat'}, inplace=True)
df_max.set_index('substat', inplace=True)

df_max = df_max[['rune_set', 'max_value']]
```
